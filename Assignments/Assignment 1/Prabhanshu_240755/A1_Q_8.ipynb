```python
import pandas as pd
import numpy as np
```

# 8.1 Loading the data file and assigning columns

```python
df = pd.read_csv('processed.cleveland.data', header = None , names = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG',
'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope', 'NumVesselsFluoro',
'Thalassemia', 'HeartDisease'])
```

```python
df
```

# 8.2.1 Replacing '?' with NaN

```python
df.replace('?' , np.nan)
```

# 8.2.2 Converting relevant columns to numeric.

```python
df['Age'] = pd.to_numeric(df['Age'], errors ='coerce')
df['RestingBP'] = pd.to_numeric(df['RestingBP'] , errors ='coerce')
df['Cholesterol'] = pd.to_numeric(df['Cholesterol'] , errors ='coerce')
df['FastingBS'] = pd.to_numeric(df['FastingBS'] , errors ='coerce')
df['RestingECG'] = pd.to_numeric(df['RestingECG'] , errors ='coerce')
df['MaxHR'] = pd.to_numeric(df['MaxHR'] , errors ='coerce')
df['ExerciseAngina'] = pd.to_numeric(df['ExerciseAngina'] , errors ='coerce')
df['Oldpeak'] = pd.to_numeric(df['Oldpeak'] , errors ='coerce')
df['ST_Slope'] = pd.to_numeric(df['ST_Slope'] , errors ='coerce')
df['NumVesselsFluoro'] = pd.to_numeric(df['NumVesselsFluoro'] , errors ='coerce')
df['Thalassemia'] = pd.to_numeric(df['Thalassemia'] , errors ='coerce')
df['HeartDisease'] = pd.to_numeric(df['HeartDisease'] , errors ='coerce')
```

# 8.3.1 Displaying the count of missing values in each column.

```python
missing_values_count = df.isnull().sum()

print(missing_values_count)
```

# 8.3.2 Replace missing values in NumVesselsFluoro and Thalassemia using the median or mode.

```python
print(df['NumVesselsFluoro'].median())
print(df['NumVesselsFluoro'].mode())
print(df['Thalassemia'].median())
print(df['Thalassemia'].mode())
```

```python
df['NumVesselsFluoro'] = df['NumVesselsFluoro'].replace(np.nan , 0.0)
```

```python
df['Thalassemia'] = df['Thalassemia'].replace(np.nan , 3.0)
```

```python
df.groupby('HeartDisease')['Cholesterol'].mean()
```
